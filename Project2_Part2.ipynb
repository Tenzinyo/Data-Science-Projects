```python
#importing relevant packages

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#loading the dataset
acc1=pd.read_csv("Acceleration 1.csv")
acc2=pd.read_csv("Acceleration 2.csv")
```

```python
acc1.head()
acc1
```

```python
#exploring the acceleration graph

acc1.plot(y='Acceleration (m/s²)')
```

```python
delta1 = acc1["Time (s)"].iloc[-1] / acc1.shape[0]
delta1
```

```python
acc1["Velocity1"] = acc1['Acceleration (m/s²)'].cumsum() * delta1
```

```python
acc1["Altitude1"] = acc1["Velocity1"].cumsum() * delta1
```

```python
acc1
```

```python
maxheight1=acc1['Altitude1'].min()
maxheight1
```

```python
acc1.plot(y='Altitude1')
```

```python
acc1.plot(y='Velocity1')
```

```python
acc2.plot(y='Acceleration (m/s²)')
```

```python
delta2 = acc2["Time (s)"].iloc[-1] / acc2.shape[0]
delta2
```

```python
acc2["Velocity2"] = acc2['Acceleration (m/s²)'].cumsum() * delta2
acc2["Altitude2"] = acc2["Velocity2"].cumsum() * delta2
acc2
```

```python

maxheight2=acc2['Altitude2'].max()
maxheight2
```

```python
acc2.plot(y='Altitude2')
```

```python
acc2.plot(y='Velocity2')
```

We measured the height of the Reid building (i.e. from the basement the second floor) operating the elevator function from Phyphox using two different iPhones. Here we used the cumsum and max function to help us figure the velocity and height.
