```python
#Task 2: Find the country that has the most number of COVID deaths
max_case=0
max_country=''
for item in res:
    url= 'https://api.covid19api.com/total/country/'+item['Slug']
   
    req=requests.get(url)        
    re=req.json()
    #print(re)
    if len(re)>2:
        case=re[-1]['Deaths']
        country = item['Country']

        if case>max_case:
            max_case=case
            max_country=country
print(max_case , max_country) 
```

```python
#Task 1: Make at least three API requests total.
sports_url= 'https://api.the-odds-api.com/v4/sports?apiKey=1915ed783eace238eb39b0e4fa150578'
sports_response= requests.get(sports_url)
sports=sports_response.json()
sports
```

```python
# Task 2: Make a request to an API and then use data from that first request to make additional requests.
#making a request to get the sport name to get a sport url 
def sport(sport_name):
    result=[]
    for item in sports:
        key=item['key']
        if sport_name in key:
            result.append(key)
    return result
```

```python
#Soccer name generation example for General URL processing 
sport('soccer')
```

```python
#getting the url using the baseurl from the API website and also the sport name generator to make a basic url
def get_url(sport):
    baseurl='https://api.the-odds-api.com/v4/sports/'
    odds='/odds/?apiKey=1915ed783eace238eb39b0e4fa150578&regions=us&markets=h2h,spreads&oddsFormat=american'
    url=baseurl + sport +odds
    return url
```
